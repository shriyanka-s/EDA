```python
# Wap ask the user get a random number between 1 to 100
# find  even or odd
# get the sum of even and sum of odd
# Get 5 random number (for 5 times)
# even_summ=0
# odd_summ=0

import random 
even = 0
odd = 0
for i in range(5):
    random_num = random.randint(1,100)
    if random_num % 2 == 0 :
        print(random_num," is an even number")
        even =+ random_num
    else:
        print(random_num,"is a odd number")
        odd =+ random_num
        
print("sum of even numbers: ",even)     
print("sum of odd numbers: ",odd)
```

```python
# Wap ask the user get a random number between 1 to 100
# find  even or odd
# get the sum of even and sum of odd
# Get 5 random number (for 5 times)
# even_summ=0
# odd_summ=0

import random 
even_sum,odd_sum = 0,0
even_count, odd_count=0,0
for i in range(5):
    random_num = random.randint(1,100)
    if random_num % 2 == 0 :
        print(random_num," is an even number")
        even_sum = even_sum + random_num
        even_count = even_count + 1
    else:
        print(random_num,"is a odd number")
        odd_sum =+ random_num
        odd_count = odd_count + 1
        
print("sum of even numbers: ",even_sum)     
print("sum of odd numbers: ",odd_sum)
print("even count: ",even_count)
print("odd count: ",odd_count)
```
