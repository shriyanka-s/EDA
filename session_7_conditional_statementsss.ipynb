```python
num1 = eval(input("Enter the first number"))
num2 = eval(input("Enter the second number"))
print("Enter from the following options \n(1) for addition \n(2) for subtraction \n(3) for multiplication \n(4) for division")

operation = int(input("Enter a number between 1 and 4"))

if operation == 1:
    add = num1 + num2
    print(f"The addition of {num1} and {num2} is: {add}")
elif operation == 2:
    sub = num1 - num2
    print(f"The subtraction of {num1} and {num2} is: {sub}")
elif operation == 3:
    div = num1 / num2
    print(f"The division of {num1} and {num2} is: {div}")
elif operation == 4:
    mul = num1 - num2
    print(f"The multiplication of {num1} and {num2} is: {mul}")
else:
    print("Enter a valid operation")
```
